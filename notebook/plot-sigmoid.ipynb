```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from matplotlib.ticker import MaxNLocator
import numpy as np
import math
import pathlib
```

```python
# necessary paths
NOTEBOOK_PATH = pathlib.Path().resolve()
IMG_DIRECTORY = NOTEBOOK_PATH.parent / "img"
```

```python
# plot sigmoid function
def sigmoid(arr, scale=1):
    result = 1/(1 + np.exp(-arr*scale))
    return result

x = np.linspace(-5, 5)
y = sigmoid(x)

fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x, y);
ax.set_xlabel('x')
ax.set_ylabel('S(x)')
ax.axhline(y=0, color='.7')
ax.axvline(x=0, color='.7');
```

```python
# save plot
plt.savefig(IMG_DIRECTORY / "sigmoid.png")
```
